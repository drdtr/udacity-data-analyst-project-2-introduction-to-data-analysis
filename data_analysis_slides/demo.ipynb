## Normal distribution

Flip a coin and plot the sample distribution in a histogram:
- n: number of coin flips in one experiment
- p: probability for heads
- size: number of experiments in the sample

```python
plt.figure(figsize=(15, 5))
plt.hist(np.random.binomial(n=20, p=0.5, size=1000000), bins=99);
```

[Normal (Gaussian) distribution](https://en.wikipedia.org/wiki/Normal_distribution): bell curve,
formula for mean value $\mu$ and standard deviation $\sigma$:

$ f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}(\frac{x - \mu}{\sigma})^2} $

Here, the Python function for the normal distribution formula:

```python
def norm_dist(x, m, sd):
    a = ((x - m) / sd) ** 2 / -2
    b = sd * np.sqrt(2 * np.pi)
    return np.exp(a) / b
```

And now we would like to plot the normal distribution curve.

```python
mean = 10
sd = 2
values = np.linspace(0, mean * 2, 101)
norm_dist_vect = np.vectorize(norm_dist)
plt.figure(figsize=(15, 5))
plt.plot(values, norm_dist_vect(values, mean, sd));
```
